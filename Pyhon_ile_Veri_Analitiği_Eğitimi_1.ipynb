### NUMPY İLE MATEMATİKSEL iŞLEMLER

```python
matris=np.arange(10,20,1)
yeni_matris=matris.reshape(5,2)
print(yeni_matris)
```



> **Matrisin tolam, minimum ve maksimum değerlerini hesaplayalım**



```python
print('Matris elemanlarının toplamı:',yeni_matris.sum())
print('Matrisin minimum elemanı:',yeni_matris.min())
print('Matris maksimum elemanı:',yeni_matris.max())
```

* **Peki matrisin satır ya da sütun elemanınlarının toplamını bulmak istersek ne yapmalıyız?**



> **Matrisin ortalama, medyan, varyans ve standart sapmasını hesaplayalım.**



```python
print('Matrisin ortalaması:', yeni_matris.mean())
print('Matrisin medyanı:', np.median(yeni_matris))
print('Matrisin varyansı:', yeni_matris.var())
print('Matrisin standart sapması:', yeni_matris.std())
```

* Burada fonksiyon kullanımlarında bir farklılık görüyormusunuz?

ndarray nesnesi np.median(ndarray) fonksiyonuna sahiptir fakat ndarray.median() metoduna sahip değildir. Metod ve fonksiyon farkı nedir??



> **Matrisimiz üzerinde toplama, çıkarma, bölme ve çarpma işlemi yapalım.**



```python
print(yeni_matris)
print('----------------------')
print('İKİ MATRİSİN TOPLAMI:')
print(yeni_matris+yeni_matris)
print('----------------------')
print('İKİ MATRİSİN FARKI:')
print(yeni_matris-yeni_matris)
print('----------------------')
print('İKİ MATRİSİN ÇARPIMI:')
print(yeni_matris*yeni_matris)
print('----------------------')
print('İKİ MATRİSİN BÖLÜMÜ:')
print(yeni_matris/yeni_matris)
print('----------------------')
```



> **Matrisler üzerinde logaritmik, üstel, trigonometrik gibi özel fonksiyon işlemleri**



```python
print(yeni_matris)
print('----------------------')
print('MATRİSİN SİNÜS DEĞERİ:')
print(np.sin(yeni_matris))
print('----------------------')
print('MATRİSİN LOGARİTMASI:')
print(np.log(yeni_matris))
print('----------------------')
print('MATRİSİN KAREKÖKÜ:')
print(np.sqrt(yeni_matris))
print('----------------------')
```



> **Matrislerin Çarpımı**
* NOT: Matris çarpımında ilk matrisin süyun sayısı ve ikinci matrisin satır sayısı eşit olmalıdır.



```python
matris1=np.arange(10,20,1)
yeni_matris1=matris1.reshape(5,2)
print(yeni_matris1)
print('----------------------')
matris2=np.arange(0,10,1)
yeni_matris2=matris2.reshape(2,5)
print(yeni_matris2)
```

```python
print('İki matrisin çarpımı: ')
print('-----------------------')
print(np.dot(yeni_matris1,yeni_matris2))
```



> **Matrisin Transpozu**



```python
print('Matrisin Transpozu:')
print('--------------------')
print(yeni_matris1.T)
```
